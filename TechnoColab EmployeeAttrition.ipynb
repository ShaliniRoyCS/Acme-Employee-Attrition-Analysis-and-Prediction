```python
# Remove the EmployeeNumber column as its unique for every employees
df.drop(columns = 'EmployeeNumber', inplace = True)
```

```python
categorcial_features = ['BusinessTravel', 'Department','Education', 'EducationField',
                       'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel', 
                        'JobRole', 'JobSatisfaction', 'MaritalStatus','NumCompaniesWorked', 'OverTime', 'PerformanceRating', 
                        'RelationshipSatisfaction', 'StockOptionLevel','TrainingTimesLastYear', 'WorkLifeBalance']

for category in categorcial_features:
    plt.figure(figsize = (8, 6))
    sns.countplot(x = category, data = df, hue = 'Attrition', palette = 'Set1')
```

```python
le = LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column]); 
```

```python
df
```

```python
df_new = df.drop(columns = ['EmployeeCount', 'StandardHours', 'Over18'])
df_new.info()
```

```python
df_new.shape
```

```python
df_new['Attrition'].value_counts()
```

```python
from scipy import stats

# Print initial number of rows
print(f'Number of rows before outlier removal: {len(df_new)}')

# Iterate over specified columns and remove outliers
for col in ['MonthlyIncome', 'NumCompaniesWorked', 'TotalWorkingYears','TrainingTimesLastYear', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']:
    
    # Calculate z-scores for the current column
    z_scores = stats.zscore(df_new[col])
    
    # Filter out rows where the absolute z-score is greater than or equal to 3
    df_new = df_new[abs(z_scores) < 3]

# Print final number of rows after outlier removal
print(f'Number of rows after outlier removal: {len(df_new)}')
```

```python
plt.figure(figsize=(25, 20))
sns.heatmap(df_new.corr(), cmap='coolwarm', annot=True, fmt=".2f", annot_kws={"size": 13})
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
# Splitting the data into features and target
x = df_new[[col for col in df_new.columns if col != 'Attrition']]  # Selecting all columns except 'Attrition' as features
y = df_new['Attrition'].values  # Selecting the 'Attrition' column as the target variable

# Print the comparison count between attrition and non-attrition classes before addressing class imbalance
print("Comparison of attrition and non-attrition classes before addressing class imbalance:")
print(pd.Series(y).value_counts())  # Counting the occurrences of each class in the 'Attrition' target variable
```

```python
# Handling class imbalance
from imblearn import over_sampling

# Apply SMOTE (Synthetic Minority Over-sampling Technique) to handle class imbalance
x_over, y_over = over_sampling.SMOTE(sampling_strategy=1).fit_resample(x, y)

# Print the comparison count between attrition and non-attrition classes after handling class imbalance
print("Comparison of attrition and non-attrition classes after handling class imbalance:")
print(pd.Series(y_over).value_counts())
```

```python
df_final = x_over
df_final['Attrition'] = y_over
df_final.info()
```
